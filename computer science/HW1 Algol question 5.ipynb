```python
def mergesort (A):
    if len(A) < 2:
        return A
    mid = len(A) // 2
    B = A[:mid]
    C = A[mid:]
    
    B= mergesort(B)
    C= mergesort(C)
    
    i=j=0
    A=[]
    while i<len(B) and j<len(C):
        if B[i] <= C[j] :
            A += [B[i]]
            i += 1
        else:
            A += [C[j]]
            j+=1
    A += B[i:] + C[j:]
    
    return (A)
```

```python
def insertionsort(A):
    for i in range(1,len(A)):
        temp = A[i]
        j= i-1
        while j >= 0 and A[j] > temp:
            A[j+1] , j = A[j] , j-1
        A[j+1] = temp
    return(A)
```
